```python
import pandas as pd
import numpy as np
```

```python
df_age = pd.read_csv("Age.csv")
print(df_age.head()) 
```

```python
df_country = pd.read_csv("Country.csv")
print(df_country.head()) 
```

```python
df_gender = pd.read_csv("Gender.csv")
print(df_gender.head()) 
```

```python
df_ytcat = pd.read_csv("YTCat.csv")
print(df_ytcat.head())  
```

```python
df_ytcon = pd.read_csv("YTCon.csv")
df_ytcon['Likes'] = df_ytcon['Likes'].str.replace(',', '').astype(int)
df_ytcon['Dislikes'] = df_ytcon['Dislikes'].str.replace(',', '').astype(int)
```

```python
df_person = pd.read_csv("Person.csv")
# Change the date format of the DOB column in the Person table
df_person['DOB'] = pd.to_datetime(df_person['DOB'], infer_datetime_format=True)
df_person['DOB'] = df_person['DOB'].dt.strftime('%Y-%m-%d')

# Change the date format of the Publish_Date column in the YTCon table
df_ytcon['Publish_date'] = pd.to_datetime(df_ytcon['Publish_date'], infer_datetime_format=True)
df_ytcon['Publish_date'] = df_ytcon['Publish_date'].dt.strftime('%Y-%m-%d')

# Save the updated DataFrames to new CSV files
df_person.to_csv("Person_clean.csv", index=False)
df_ytcon.to_csv("YTCon_clean.csv", index=False)
```

```python
# Iterate over each row in the Person DataFrame
for i, row in df_person.iterrows():
    # Get the Person_ID and Initial_Name values for the current row
    person_id = row['Person_ID']
    initial_name = row['Initial_Name']

    # Find the corresponding row in the Gender DataFrame
    gender_row = df_gender[df_gender['Person_ID'] == person_id]

    # Update the Gender_code value based on the Initial_Name value
    if initial_name.startswith("Mrs") or initial_name.startswith("Ms.") or initial_name.startswith("Ms") or initial_name.startswith("Mrs."):
        df_gender.loc[gender_row.index, 'Gender_code'] = 'F'
    
    elif initial_name.startswith("Mr") or initial_name.startswith("Mr."):
        df_gender.loc[gender_row.index, 'Gender_code'] = 'M'
    else:
        df_gender.loc[gender_row.index, 'Gender_code'] = 'UK'

df_gender['Gender_dsc'] = np.where(df_gender['Gender_code'] == 'M', 'Male',
                                   np.where(df_gender['Gender_code'] == 'F', 'Female', 'Unknown'))

# Save the updated Gender DataFrame to a new CSV file
df_gender.to_csv("Gender_clean.csv", index=False)
```

```python
df_ytcon['Min_age'] = np.where(df_ytcon['Min_age'] < 13, 'U13',
                                     np.where((df_ytcon['Min_age'] >= 13) & (df_ytcon['Min_age'] < 18), 'U18',
                                     np.where((df_ytcon['Min_age'] >= 18) & (df_ytcon['Min_age'] < 21), 'U21',
                                              'T21')))
df_ytcon.to_csv("YTcon_clean.csv", index=False)
```
